```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler,LabelEncoder
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Activation,Dropout
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.callbacks import EarlyStopping
from sklearn.metrics import classification_report,confusion_matrix
```

```python
df=pd.read_csv('iris.csv')
```

```python
df
```

```python
df.isnull().sum()
```

```python
x=df.drop('species',axis=1).values.astype(float)
y=df['species'].values
```

```python
x
```

```python
y
```

```python
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
X_train.shape
```

```python

X_test.shape
```

```python
y_test.shape
```

```python
y_train.shape
```

label encoding followed by one hot encoding

```python
label_encoder = LabelEncoder()
label_encoder.fit(y_train)
encoded_y_train = label_encoder.transform(y_train)
print(encoded_y_train)
y_train=to_categorical(encoded_y_train)

```

```python
#y test 
label_encoder.fit(y_test)
encoded_y_test = label_encoder.transform(y_test)
#one hot encoded)
y_test=to_categorical(encoded_y_test)
```

```python
y_train
```

```python
y_test
```

```python
scaler=MinMaxScaler()
X_train=scaler.fit_transform(X_train)
X_test=scaler.transform(X_test)
```

designing the model

```python
early_stop=EarlyStopping(monitor='val_loss',patience=20)
```

```python
model=Sequential()
model.add(Dense(40,activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(20,activation='relu'))
model.add(Dropout(0.5))


model.add(Dense(3,activation='softmax'))


model.compile(optimizer='adam',loss='categorical_crossentropy')
```

```python
model.fit(X_train,y_train,epochs=600,validation_data=(X_test,y_test),callbacks=[early_stop])
```

```python
pd.DataFrame(model.history.history).plot()
```

```python
predictions=model.predict_classes(X_test, batch_size=15, verbose=0)
```

```python
predictions
```

```python
y_test
```

```python
print(classification_report(encoded_y_test,predictions))
```

```python
print(confusion_matrix(encoded_y_test,predictions))
```
